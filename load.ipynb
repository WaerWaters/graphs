```python
import os
import geopandas as gpd

data_folder = './data'
all_data = {}

for filename in os.listdir(data_folder):
    if filename.endswith('.gpkg'):
        file_path = os.path.join(data_folder, filename)
        try:
            data = gpd.read_file(file_path)
            filename_without_ext = os.path.splitext(filename)[0]
            cleaned_filename = filename_without_ext.replace('network', '').replace('[', '').replace(']', '').replace('\'', '')
            all_data[cleaned_filename] = data
        except Exception as e:
            print(f"Failed to read file {file_path}: {e}")
```

DictData

```python
import os
import pickle

data_folder = './dictData'
all_data = {}

for filename in os.listdir(data_folder):
    if filename.endswith('.pickle'):
        file_path = os.path.join(data_folder, filename)
        try:
            with open(file_path, 'rb') as handle:
                data = pickle.load(handle)["['San Francisco, California, USA']"]
                filename_without_ext = os.path.splitext(filename)[0]
                cleaned_filename = filename_without_ext.replace('[', '').replace(']', '').replace('\'', '')
                all_data[cleaned_filename] = data
        except Exception as e:
            print(f"Failed to read file {file_path}: {e}")
```
